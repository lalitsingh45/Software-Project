```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
import time
import pandas

s = Service(r"chromedriver.exe")
driver = webdriver.Chrome(service=s)

driver.get('https://vitchennaievents.com/conf1/')
time.sleep(2)
cards = driver.find_elements("xpath", "/html/body/center[2]/div/div")
title = []
description = []
for c in cards:
    if c.is_displayed() and c.is_enabled():
        # click on the event
        c.click()
        time.sleep(5)
    else:
        continue
    tl = driver.find_element('xpath', '/html/body/div[4]/div/div/div[2]/table/tbody/tr[1]/td')
    title.append(tl.text)
    time.sleep(1)
    driver.find_element('xpath', '/html/body/div[4]/div/div/div[2]/table/tbody/tr[2]/td/div/p/i').click()
    time.sleep(3)
    desc = driver.find_element('xpath', '/html/body/div[4]/div')
    description.append(desc.text)
    time.sleep(2)
    driver.find_element('xpath', '/html/body/div[4]/div/button').click()
    time.sleep(2)
    driver.find_element('xpath', '/html/body/div[4]/div/div/div[2]/span').click()

print(len(title))
print(len(description))

df = pandas.DataFrame({'Event name': title, 'Description': description})
df.to_excel('Event details123.xlsx')
print('successful')
```
